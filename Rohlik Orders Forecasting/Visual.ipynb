```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
train = pd.read_csv(r'C:\Users\MONSTER\Desktop\train.csv')

train['date'] = pd.to_datetime(train['date'])
train.set_index('date', inplace=True)
```

```python
daily_orders = train.resample('D')['orders'].sum()

# Zaman serisi grafii
plt.figure(figsize=(14, 7))
plt.plot(daily_orders.index, daily_orders, marker='o', linestyle='-')
plt.title('Gnlk Sipari Says Zaman Serisi')
plt.xlabel('Tarih')
plt.ylabel('Sipari Says')
plt.grid(True)
plt.show()
```

```python


# train.csv dosyasn ykleme
train = pd.read_csv(r'C:\Users\MONSTER\Desktop\train.csv')

# Tarih stununu datetime formatna evirme
train['date'] = pd.to_datetime(train['date'])

# Depoya gre gnlk ortalama sipari saylarn hesaplama
daily_orders = train.groupby(['warehouse', 'date'])['orders'].sum().reset_index()

# rnek bir depo iin zaman serisi grafii
warehouse_name = 'Depo_A'  # Deitirilebilir, istediiniz depo ad
data = daily_orders[daily_orders['warehouse'] == warehouse_name]

plt.figure(figsize=(12, 6))
plt.plot(data['date'], data['orders'], marker='o', linestyle='-')
plt.title(f'Zaman Serisi - Gnlk Sipariler ({warehouse_name})')
plt.xlabel('Tarih')
plt.ylabel('Sipari Says')
plt.grid(True)
plt.tight_layout()
plt.show()

# Tatil gnlerindeki sipari saylarn gsteren ubuk grafii
holiday_orders = train[train['holiday'] == 1]

plt.figure(figsize=(10, 6))
sns.barplot(x='holiday_name', y='orders', data=holiday_orders)
plt.title('Tatil Gnlerinde Sipari Saylar')
plt.xlabel('Tatil Ad')
plt.ylabel('Ortalama Sipari Says')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
