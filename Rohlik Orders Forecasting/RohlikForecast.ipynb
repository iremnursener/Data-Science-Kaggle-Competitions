### NOW ADD SOME INDICATORS DIRECTLIY RELATED WITH ORDERS

```python
import pandas as pd
import numpy as np

def add_indicator(df):
    df = df.copy()

    # Exponential Moving Averages (EMA)
    df['EMA_12'] = df['orders'].shift(1).ewm(span=12, adjust=False).mean()
    df['EMA_26'] = df['orders'].shift(1).ewm(span=26, adjust=False).mean()
    df['MACD'] = df['EMA_12'] - df['EMA_26']
    df['Signal_Line'] = df['MACD'].shift(1).ewm(span=9, adjust=False).mean()

    # Bollinger Bands
    df['20_SMA'] = df['orders'].shift(1).rolling(window=20).mean()
    df['20_STD'] = df['orders'].shift(1).rolling(window=20).std()
    df['Upper_Band'] = df['20_SMA'] + (2 * df['20_STD'])
    df['Lower_Band'] = df['20_SMA'] - (2 * df['20_STD'])
    
    # Relative Strength Index (RSI)
    delta = df['orders'].shift(1).diff(1)
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    avg_gain = gain.rolling(window=14).mean()
    avg_loss = loss.rolling(window=14).mean()
    rs = avg_gain / avg_loss
    df['RSI'] = 100 - (100 / (1 + rs))

    return df
```

### TRY BUNCH OF MODELS

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor


# KullanÄ±lacak modellerin listesi
models = {
    'Linear Regression': LinearRegression(),
    'Ridge': Ridge(),
    'Lasso': Lasso(),
    'ElasticNet': ElasticNet(),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(),
    'Gradient Boosting': GradientBoostingRegressor(),
    'AdaBoost': AdaBoostRegressor(),
    'Support Vector Regressor': SVR(),
    'K-Nearest Neighbors': KNeighborsRegressor(),
    'XGBoost': XGBRegressor(),
    'CatBoost': CatBoostRegressor(verbose=0),
    'LightGBM': LGBMRegressor()
}

# MAPE hesaplamasÄ± ve sonuÃ§larÄ± saklama
results = {}

for name, model in models.items():
    model.fit(X_train.fillna(X_train.mean()), y_train)
    predictions = model.predict(X_test.fillna(X_test.mean()))
    mape = mean_absolute_percentage_error(y_test, predictions)
    results[name] = mape
    print(f"{name}: MAPE = {mape:.4f}")

# SonuÃ§larÄ± DataFrame olarak gÃ¶sterme
results_df = pd.DataFrame(list(results.items()), columns=['Model', 'MAPE'])
results_df = results_df.sort_values(by='MAPE', ascending=True)
print("\nModel PerformanslarÄ±:")
print(results_df)
```
