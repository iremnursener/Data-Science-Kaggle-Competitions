```python
def create_sequences(data, target, window_size):
    X = []
    y = []
    for i in range(len(data) - window_size):
        X.append(data[i:i + window_size])
        y.append(target[i + window_size])
    return np.array(X), np.array(y)
```

```python
# Train-test split
train_size = int(len(train_prague_1) * 0.8)
train_features, test_features = train_prague_1[:train_size], train_prague_1[train_size:]
train_target, test_target = prague_y[:train_size], prague_y[train_size:]

# Train seti için zaman serisi pencere oluşturma
train_window_size = 7  # Örnek olarak 30 günlük bir pencere
X_train, y_train = create_sequences(train_features.values, train_target.values, train_window_size)

# Test seti için zaman serisi pencere oluşturma
test_window_size = 7  # Train setiyle aynı pencere boyutunu kullanıyoruz
X_test, y_test = create_sequences(test_features.values, test_target.values, test_window_size)
```

```python
import pandas as pd
from ta.trend import MACD

def calculate_macd_with_signals(prices, window_slow=26, window_fast=12, window_sign=9):
    """
    Son 26 kapanış fiyatından MACD, sinyal hattı ve alım/satım sinyallerini hesaplar.

    Parametreler:
    prices (list or pd.Series): Son 26 kapanış fiyatı.
    window_slow (int): Uzun dönem EMA penceresi.
    window_fast (int): Kısa dönem EMA penceresi.
    window_sign (int): Sinyal hattı EMA penceresi.

    Geri dönüş:
    dict: MACD, sinyal hattı, MACD histogramı ve alım/satım sinyallerini içeren sözlük.
    """

    if len(prices) < window_slow:
        raise ValueError("MACD hesaplamak için en az 26 kapanış fiyatına ihtiyacınız var.")

    df = pd.DataFrame(prices.values,columns=["Close"])
    
    macd_indicator = MACD(close=df['Close'], window_slow=window_slow, window_fast=window_fast, window_sign=window_sign)
    df['MACD'] = macd_indicator.macd()
    df['MACD_Signal'] = macd_indicator.macd_signal()
    df['MACD_Histogram'] = macd_indicator.macd_diff()

    # Alım/Satım sinyallerini oluşturma
    df['Signal'] = 0  # Varsayılan olarak sinyal yok
    df.loc[df['MACD'] > df['MACD_Signal'], 'Signal'] = 1
    df.loc[df['MACD'] < df['MACD_Signal'], 'Signal'] = -1

    return df.iloc[-1][df.columns[1:]]

calculate_macd_with_signals(real_y[-34:])
```

```python
import pandas as pd
from ta.volatility import BollingerBands

def calculate_rsi(data, window=14):
    """
    RSI (Relative Strength Index) hesaplar.
    
    Parametreler:
    data (pd.Series): Kapanış fiyatları.
    window (int): RSI hesaplaması için kullanılan pencere boyutu.
    
    Geri dönüş:
    pd.Series: RSI değerleri.
    """
    delta = data.diff()

    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)

    avg_gain = gain.rolling(window=window, min_periods=1).mean()
    avg_loss = loss.rolling(window=window, min_periods=1).mean()

    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))

    return rsi

def calculate_indicators(prices, rsi_window=14, bb_window=20, bb_window_dev=2):
    """
    Son kapanış fiyatlarından RSI ve Bollinger Bantlarını hesaplar.
    
    Parametreler:
    prices (list or pd.Series): Kapanış fiyatları.
    rsi_window (int): RSI hesaplaması için pencere boyutu.
    bb_window (int): Bollinger Bantları hesaplaması için pencere boyutu.
    bb_window_dev (int): Bollinger Bantları için standart sapma.
    
    Geri dönüş:
    dict: RSI, Bollinger Bantları ve kapanış fiyatlarını içeren sözlük.
    """
    
    if len(prices) < max(rsi_window, bb_window):
        raise ValueError("Yeterli kapanış fiyatı yok. En az {} kapanış fiyatına ihtiyacınız var.".format(max(rsi_window, bb_window)))

    df = pd.DataFrame(prices, columns=['Close'])

    # RSI hesaplama
    df['RSI'] = calculate_rsi(df['Close'], window=rsi_window)

    # Bollinger Bantları hesaplama
    bb_indicator = BollingerBands(close=df['Close'], window=bb_window, window_dev=bb_window_dev)
    df['BB_Middle'] = bb_indicator.bollinger_mavg()
    df['BB_Upper'] = bb_indicator.bollinger_hband()
    df['BB_Lower'] = bb_indicator.bollinger_lband()

    return {
        
        'BB_Middle': df['BB_Middle'].iloc[-1],
        'BB_Upper': df['BB_Upper'].iloc[-1],
        'BB_Lower': df['BB_Lower'].iloc[-1],
        'RSI': df['RSI'].iloc[-1]
    }

# Örnek kullanım
prices = [150, 152, 153, 154, 155, 156, 157, 158, 159, 160, 161, 162, 163, 164, 165, 166, 167, 168, 169, 170, 171, 172, 173, 174, 175, 176]
result = calculate_indicators(real_y[-34:].values)
print(result)
calculate_rsi(real_y[-34:])
```
