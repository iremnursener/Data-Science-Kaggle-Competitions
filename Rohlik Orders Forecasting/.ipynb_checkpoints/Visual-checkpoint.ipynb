```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
train = pd.read_csv(r'C:\Users\MONSTER\Desktop\train.csv')

train['date'] = pd.to_datetime(train['date'])
train.set_index('date', inplace=True)
```

```python
daily_orders = train.resample('D')['orders'].sum()

# Zaman serisi grafiği
plt.figure(figsize=(14, 7))
plt.plot(daily_orders.index, daily_orders, marker='o', linestyle='-')
plt.title('Günlük Sipariş Sayısı Zaman Serisi')
plt.xlabel('Tarih')
plt.ylabel('Sipariş Sayısı')
plt.grid(True)
plt.show()
```

```python


# train.csv dosyasını yükleme
train = pd.read_csv(r'C:\Users\MONSTER\Desktop\train.csv')

# Tarih sütununu datetime formatına çevirme
train['date'] = pd.to_datetime(train['date'])

# Depoya göre günlük ortalama sipariş sayılarını hesaplama
daily_orders = train.groupby(['warehouse', 'date'])['orders'].sum().reset_index()

# Örnek bir depo için zaman serisi grafiği
warehouse_name = 'Depo_A'  # Değiştirilebilir, istediğiniz depo adı
data = daily_orders[daily_orders['warehouse'] == warehouse_name]

plt.figure(figsize=(12, 6))
plt.plot(data['date'], data['orders'], marker='o', linestyle='-')
plt.title(f'Zaman Serisi - Günlük Siparişler ({warehouse_name})')
plt.xlabel('Tarih')
plt.ylabel('Sipariş Sayısı')
plt.grid(True)
plt.tight_layout()
plt.show()

# Tatil günlerindeki sipariş sayılarını gösteren çubuk grafiği
holiday_orders = train[train['holiday'] == 1]

plt.figure(figsize=(10, 6))
sns.barplot(x='holiday_name', y='orders', data=holiday_orders)
plt.title('Tatil Günlerinde Sipariş Sayıları')
plt.xlabel('Tatil Adı')
plt.ylabel('Ortalama Sipariş Sayısı')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
train['date'] = pd.to_datetime(train['date'])
plt.figure(figsize=(15, 5))  # Figür boyutunu ayarlayabilirsiniz
plt.scatter(train['date'], train['user_activity_2'], color='blue', alpha=0.6, s=15, label='User Activity')  # Nokta grafiği oluşturma
plt.xlabel('Date')
plt.ylabel('User Activity')
plt.title('User Activity Over Time')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.4)  # Izgara ekleyerek arka planı görsel olarak bölme
plt.tight_layout()

# X eksenindeki tarih etiketlerinin sıklığını ve formatını ayarla (örneğin, her 3 ayda bir)
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=3))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

plt.xticks(rotation=45)  # Tarih etiketlerini 45 derece döndürme (isteğe bağlı)
plt.gca().patch.set_facecolor('lightgray')  # Arka plan rengini açık gri yapma

plt.show()
```
